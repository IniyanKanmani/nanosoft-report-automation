```python
import pandas as pd
import numpy as np
import openpyxl as el
```

```python
input_file = './Transaction-Mar-01-2024-02_30_18-0700-12157472268.csv'
output_file = './Ebay Manual template.xlsx'
```

```python
data = pd.read_csv(input_file, header=11)
```

```python
data = data.replace('--', np.float64(0.0))
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.rename(columns={'Final value fee \u2013 fixed': 'Final value fee - fixed', 'Final value fee \u2013 variable': 'Final value fee - variable'})
```

```python
def convert_to_numeric(data, column_name):
    if data[column_name].dtype != np.float64:
        for i in range(len(data)):
            if data[column_name].dtype != None or type(data[column_name]) != float:
                try:
                    data[column_name][i] = float(data[column_name][i].replace(',', ''))
                except:
                    print(data[column_name][i], type(data[column_name][i]))
        data[column_name] = pd.to_numeric(data[column_name])
    return data
```

```python
data = convert_to_numeric(data, 'Gross transaction amount')
data = convert_to_numeric(data, 'Final value fee - fixed')
data = convert_to_numeric(data, 'Final value fee - variable')
data = convert_to_numeric(data, "Very high 'item not as described' fee")
data = convert_to_numeric(data, "Below standard performance fee")
```

```python
uk_sales_gross = data.loc[data['Type'] == 'Order']['Gross transaction amount'].astype(np.float64).sum()
```

```python
sales_fees_df = data.loc[data['Type'] == 'Order'][['Final value fee - fixed', 'Final value fee - variable', 'Very high \'item not as described\' fee', 'Below standard performance fee']]
```

```python
sales_fees_df['total sales fee'] = -np.inf
```

```python
for i in sales_fees_df.index:
    sales_fees_df['total sales fee'][i] = sales_fees_df['Final value fee - fixed'][i] + sales_fees_df['Final value fee - variable'][i] + sales_fees_df['Very high \'item not as described\' fee'][i] + sales_fees_df['Below standard performance fee'][i]
```

```python
sales_fees = sales_fees_df['total sales fee'].astype(np.float64).sum() * -1
```

```python
other_fees = data.loc[data['Type'] == 'Other fee']['Gross transaction amount'].astype(np.float64).sum() * -1
```

```python
wb = el.load_workbook(output_file)
```

```python
ws = wb.active
```

```python
ws['B4'].value = uk_sales_gross
```

```python
ws['B9'].value = sales_fees
```

```python
ws['B10'].value = other_fees
```

```python
wb.save(output_file)
```
